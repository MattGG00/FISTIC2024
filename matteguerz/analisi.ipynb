```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('house.csv', index_col=0)
df
```

```python
# Creo le variabili da inserire nel grafico
X = df[['LSTAT','RM']]          # Reshaping necessario per una singola feature
y = df['Price']                 # variabile dipendente, target

# divido i db per Train & Test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)

# Creazione regressione
model = LinearRegression()

# adatto il modello ai dati
model.fit(X_train, y_train)

# controllo bontÃ  del modello
y_pred = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

# intercetta e coefficiente
coefficienti = model.coef_      # beta regressori
intercetta = model.intercept_   # punto partenza della retta

# printo i risultati
print(f'root_mean_squared_error: {rmse}')
print(f'r2: {r2}')
print(f'coefficiente: {coefficienti}')
print(f'intercetta: {intercetta}')

# Creazione della griglia per la superficie di regressione
# np.meshgrid crea una griglia 2D di punti combinando i valori di LSTAT e RM
# np.linspace crea 20 punti equidistanti tra il min e max di ogni variabile
x1_grid, x2_grid = np.meshgrid(
   np.linspace(X_test['LSTAT'].min(), X_test['LSTAT'].max(), 20),  # 20 punti per LSTAT
   np.linspace(X_test['RM'].min(), X_test['RM'].max(), 20)         # 20 punti per RM
)

# Creiamo un DataFrame per la griglia
# ravel() appiattisce le griglie 2D in array 1D
# column_stack combina gli array in una matrice di due colonne
# Assegniamo i nomi delle colonne originali per compatibilitÃ  con il modello
grid_points = pd.DataFrame(
   np.column_stack((x1_grid.ravel(), x2_grid.ravel())),
   columns=['LSTAT', 'RM']
)

# Facciamo le previsioni sulla griglia di punti
# reshape riporta il risultato alla forma originale della griglia (20x20)
y_grid = model.predict(grid_points).reshape(x1_grid.shape)

# Inizializziamo una figura plotly vuota
fig = go.Figure()

# Aggiungiamo i punti di scatter (dati reali)
# Scatter3d crea un grafico a dispersione 3D
fig.add_trace(go.Scatter3d(
   x=X_test['LSTAT'],     # Valori LSTAT sull'asse x
   y=X_test['RM'],        # Valori RM sull'asse y
   z=y_test,              # Valori Price sull'asse z
   mode='markers',        # Tipo di visualizzazione: punti
   marker=dict(
       size=5,            # Dimensione dei punti
       color='blue',      # Colore dei punti
       opacity=0.8        # Trasparenza dei punti (0=trasparente, 1=opaco)
   ),
   name='Dati reali'      # Nome nella legenda
))

# Aggiungiamo la superficie di regressione
# Surface crea una superficie 3D che rappresenta le previsioni del modello
fig.add_trace(go.Surface(
   x=x1_grid,            # Griglia di valori LSTAT
   y=x2_grid,            # Griglia di valori RM
   z=y_grid,             # Valori previsti dal modello
   name='Piano di regressione',
   opacity=0.7,          # Trasparenza della superficie
   colorscale='Reds'     # Schema colori della superficie
))

# Personalizzazione del layout del grafico
fig.update_layout(
   title='Regressione Lineare Multipla 3D Interattiva',  # Titolo del grafico
   scene=dict(
       xaxis_title='LSTAT',    # Etichetta asse x
       yaxis_title='RM',       # Etichetta asse y
       zaxis_title='Price'     # Etichetta asse z
   ),
   width=1000,                 # Larghezza del grafico in pixel
   height=800,                 # Altezza del grafico in pixel
   autosize=True,             # Permette il ridimensionamento automatico
   showlegend=True            # Mostra la legenda
)

# Visualizza il grafico
# show() renderÃ  il grafico interattivo nel notebook
fig.show()
```

```python
# visualizzo errore tra 'y reale' e 'y pred'
lenght = y_pred.shape[0]  # prende le righe di 'y_pred'
x = np.linspace(0, lenght, lenght)  #la x (ascisse) viene costruita partendo dalla riga zero (0) fino alla fine delle righe (primo 'lenght') allineandomi tutti i punti della lunghezza 'lenght' (secondo 'lenght')

plt.figure(figsize = (10, 7))
plt.plot(x, y_test, label = 'real y')
plt.plot(x, y_pred, label = 'pred y')
plt.legend()
```
