```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns

import datetime
```

```python
nvda = yf.Ticker("NVDA")
tsla = yf.Ticker("TSLA")
appl = yf.Ticker("AAPL")
```

## NVDA DF

```python
df_nvda = pd.DataFrame(nvda.history(start = "2000-01-01", end = "2025-01-21", actions = False))
df_nvda
```

```python
title = 'Yearly Mean Closing Price for NVDA'
df_nvda['Close'].resample('Y').mean().plot(title = title,
                                           figsize=(10, 6),
                                           color = 'blue');
```

```python
title = 'Monthly Mean Closing Price for NVDA (2024)'
df_nvda['Close']['2024'].resample('M').mean().plot.bar(title = title,
                                           figsize=(10, 6),
                                           color = 'red');
```

## TSLA DF

```python
df_tsla = pd.DataFrame(tsla.history(start = "2000-01-01", end = "2025-01-21", actions = False))
df_tsla
```

```python
title = '2024 monhtly mean Closing Price for TSLA'
df_tsla['Close']['2024'].resample('M').mean().plot.bar(title = title,
                                           figsize=(10, 6),
                                           color = 'green');
```
