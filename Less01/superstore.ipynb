```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import pygwalker as pyg
```

```python
# Assicurati che la colonna 'data' sia in formato datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Filtra il DataFrame per mantenere solo le righe tra il 1/1/2014 e il 31/12/2016
filtered_df = df[(df['Order Date'] >= '2014-01-01') & (df['Order Date'] <= '2016-12-31')]

# Mostra il DataFrame filtrato
print(filtered_df)
```

```python
def analyze_sales_by_category_region(df):
    """
    Analizza e visualizza le vendite per categoria e regione
    utilizzando i valori della colonna Sales
    
    Parameters:
    df: DataFrame con colonne 'Order Date', 'Region', 'Category', 'Sales'
    """
    # Raggruppa i dati per Regione e Categoria e somma le vendite
    sales_by_region_cat = df.groupby(['Region', 'Category'])['Sales'].sum().unstack()

    # Crea la figura e definisci le dimensioni
    plt.figure(figsize=(12, 6))

    # Crea il grafico a barre raggruppate
    ax = sales_by_region_cat.plot(kind='bar', width=0.8)

    # Personalizza il grafico
    plt.title('Vendite Totali per Categoria e Regione (2014-2016)', fontsize=14, pad=20)
    plt.xlabel('Regione', fontsize=12)
    plt.ylabel('Denaro tot. speso per categoria', fontsize=12)
    plt.legend(title='Categoria', bbox_to_anchor=(1.05, 1), loc='upper left')

    # Aggiungi la griglia
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Ruota le etichette dell'asse x per una migliore leggibilità
    plt.xticks(rotation=45, ha='right')

    # Aggiusta il layout per evitare il taglio delle etichette
    plt.tight_layout()

    # Mostra il grafico
    plt.show()

    return sales_by_region_cat

# Ora puoi chiamare la funzione con il tuo DataFrame già filtrato:
sales_data = analyze_sales_by_category_region(df)
```

```python
def analyze_quantity_by_region(df):
    """
    Analizza e visualizza la quantità di prodotti ordinati per regione nel periodo 2014-2016
    
    Parameters:
    df: DataFrame con colonne 'Order Date', 'Region', 'Quantity'
    """
    # Crea una colonna per il mese e anno
    df['Month-Year'] = df['Order Date'].dt.to_period('M')
    
    # Raggruppa i dati per mese-anno e Regione, somma le quantità
    quantity_by_region = df.groupby(['Month-Year', 'Region'])['Quantity'].sum().unstack()
    
    # Converti l'indice in stringa per una migliore visualizzazione
    quantity_by_region.index = quantity_by_region.index.astype(str)
    
    # Crea la figura e definisci le dimensioni
    plt.figure(figsize=(12, 7))
    
    # Crea il grafico a linee
    ax = quantity_by_region.plot(kind='line', marker='o', markersize=4)
    
    # Personalizza il grafico
    plt.title('Quantità prodotti ordinati per Regione (2014-2016)', fontsize=14, pad=20)
    plt.xlabel('Tempo', fontsize=12)
    plt.ylabel('Quantità Richiesta', fontsize=12)
    
    # Posiziona la legenda sotto il grafico
    plt.legend(title='Regione', loc='upper center', bbox_to_anchor=(0.5, -0.45), ncol=4)
    
    # Aggiungi la griglia
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Ruota le etichette dell'asse x per una migliore leggibilità
    plt.xticks(rotation=45, ha='right')
    
    # Aggiusta il layout per evitare il taglio delle etichette
    plt.tight_layout()
    
    # Mostra il grafico
    plt.show()
    
    return quantity_by_region

# Ora puoi chiamare la funzione con il tuo DataFrame già filtrato:
quantity_data = analyze_quantity_by_region(df)
```
