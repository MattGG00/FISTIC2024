```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
df = pd.read_excel('https://frenzy86.s3.eu-west-2.amazonaws.com/python/data/realcase/dataset_costi.xlsx')

pd.set_option('display.max_columns', None)

df
```

# EDA

```python
df.sort_values(by=df.columns.tolist(), ascending=True)
```

```python
df = df.drop(columns=['index', 'Ragione sociale', 'ATECO 2007\ncodice', 'Indirizzo sede legale', 'Indirizzo sede legale - Codice postale','Indirizzo sede legale - Comune', 'Provincia', 'Indirizzo sede legale - Regione', 'Numero CCIAA', 'Codice fiscale', 'Partita IVA'], inplace=False)
```

```python
# Make a copy of the dataframe
df_clean = df.copy()

# First convert all 'n.d' to NaN
df_clean = df_clean.replace('n.d', np.nan)

# Convert all columns to numeric where possible
for col in df_clean.columns:
    try:
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
    except:
        print(f"Column {col} could not be converted to numeric")
        continue

# Replace NaN with median for each column
for col in df_clean.columns:
    median_val = df_clean[col].median()
    df_clean[col].fillna(median_val, inplace=True)
    print(f"Column {col}: replaced NaN values with median = {median_val}")

df_clean
```

```python
target = 'costi anno 2020'
correlations = df_clean.corr()[target].abs()

# Sort correlations by absolute value
correlations_sorted = correlations.sort_values(ascending=False)

# Get columns with correlation < 0.5, excluding the target variable itself
columns_to_keep = correlations[correlations < 0.5].index.tolist()

# Remove target from columns_to_keep if it's there (it would have correlation = 1 with itself)
if target in columns_to_keep:
    columns_to_keep.remove(target)

# Add target variable back to the list
columns_to_keep.append(target)

# Create filtered DataFrame with only the selected columns
df_corr = df_clean[columns_to_keep]
```

```python
df_corr
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Creo le variabili da inserire nel grafico
X = df_clean[['var1', 'var2', 'var93', 'var94']]  # Variabili indipendenti
y = df_clean['costi anno 2020']  # Variabile dipendente, target

# Divido i dati in Train & Test (80% training, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)

# Creazione regressione
model = LinearRegression()

# Adatto il modello ai dati
model.fit(X_train, y_train)

# Controllo la bont del modello
y_pred = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

# Intercetta e coefficienti (Beta regressori)
coefficienti = model.coef_      # Beta regressori
intercetta = model.intercept_   # Punto di partenza della retta

# Stampo i risultati
print(f'root_mean_squared_error: {rmse}')
print(f'r2: {r2}')
print(f'coefficiente: {coefficienti}')
print(f'intercetta: {intercetta}')

# ---------------------------- Creazione della griglia per la superficie di regressione ----------------------------

# Creiamo una griglia per le variabili che useremo per il grafico (due variabili per il grafico 3D)
# Usando var1, var2 e var93 per il grafico 3D
x1_grid, x2_grid = np.meshgrid(
    np.linspace(X_test['var1'].min(), X_test['var1'].max(), 20),  # 20 punti per var1
    np.linspace(X_test['var2'].min(), X_test['var2'].max(), 20)   # 20 punti per var2
)

# Creiamo un DataFrame per la griglia
# ravel() appiattisce le griglie 2D in array 1D
# column_stack combina gli array in una matrice di due colonne
# Assegniamo i nomi delle colonne originali per compatibilit con il modello
grid_points = pd.DataFrame(
   np.column_stack((x1_grid.ravel(), x2_grid.ravel(), np.full_like(x1_grid.ravel(), X_test['var93'].mean()), np.full_like(x1_grid.ravel(), X_test['var94'].mean()))),
   columns=['var1', 'var2', 'var93', 'var94']
)

# Facciamo le previsioni sulla griglia di punti
# Reshape riporta il risultato alla forma originale della griglia (20x20)
y_grid = model.predict(grid_points).reshape(x1_grid.shape)

# Inizializziamo una figura plotly vuota
fig = go.Figure()

# Aggiungiamo i punti di scatter (dati reali)
fig.add_trace(go.Scatter3d(
   x=X_test['var1'],      # Valori var1 sull'asse x
   y=X_test['var2'],      # Valori var2 sull'asse y
   z=y_test,              # Valori Price sull'asse z
   mode='markers',        # Tipo di visualizzazione: punti
   marker=dict(
       size=5,            # Dimensione dei punti
       color='blue',      # Colore dei punti
       opacity=0.8        # Trasparenza dei punti (0=trasparente, 1=opaco)
   ),
   name='Dati reali'      # Nome nella legenda
))

# Aggiungiamo la superficie di regressione
fig.add_trace(go.Surface(
   x=x1_grid,            # Griglia di valori var1
   y=x2_grid,            # Griglia di valori var2
   z=y_grid,             # Valori previsti dal modello
   name='Piano di regressione',
   opacity=0.7,          # Trasparenza della superficie
   colorscale='Reds'     # Schema colori della superficie
))

# Personalizzazione del layout del grafico
fig.update_layout(
   title='Regressione Lineare Multipla 3D Interattiva',  # Titolo del grafico
   scene=dict(
       xaxis_title='var1',    # Etichetta asse x
       yaxis_title='var2',    # Etichetta asse y
       zaxis_title=''    # Etichetta asse z
   ),
   width=1000,                 # Larghezza del grafico in pixel
   height=800,                 # Altezza del grafico in pixel
   autosize=True,             # Permette il ridimensionamento automatico
   showlegend=True            # Mostra la legenda
)

# Visualizza il grafico
fig.show()
```
