```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
# !wget https://frenzy86.s3.eu-west-2.amazonaws.com/python/data/cars.csv
path = 'https://frenzy86.s3.eu-west-2.amazonaws.com/python/data/cars.csv'

df = pd.read_csv(path)
df
```

```python
df['fuel_type'] = df['fuel_type'].fillna(df['fuel_type'].mode()[0])
```

```python
df
```

```python
df['kms_driven'] = df['kms_driven'].str.replace('kms', '').str.replace('Petrol', '').str.replace(',', '')
```

```python
df['kms_driven'] = pd.to_numeric(df['kms_driven'], errors='coerce')         # 2. Conversione in numeri

df['kms_driven'] = df['kms_driven'].fillna(df['kms_driven'].median())       # 3. fill NaN
```
